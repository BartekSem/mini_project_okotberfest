```python
import pandas as pd
```

```python
df = pd.read_csv("merged_df.csv")
df.head()
```

```python
# duration and beer consumption
correlation_duration = df['Duration'].corr(df['Beer_consumption'])

# visitors total and beer consumption
correlation_visitors = df['Visitors_total'].corr(df['Beer_consumption'])

# goals and beer consumption
correlation_goals = df['Goals'].corr(df['Beer_consumption'])

# chicken consumption and beer consumption
correlation_chicken = df['Chicken_consumption'].corr(df['Beer_consumption'])

# Beer price and beer consumption
correlation_beer_price = df['Beer_price'].corr(df['Beer_consumption'])

# inflation rate and beer consumption
correlation_inflation = df[' Inflation_Rate'].corr(df['Beer_consumption'])


print("CORRELATION ANALYSIS ")
print("--------------------------------------------------")
print(f"beer consumption beer price: {correlation_beer_price}")
print(f"beer consumption duration: {correlation_duration}")
print(f"beer consumption goals: {correlation_goals}")
print(f"beer consumption inflation: {correlation_inflation}")
print(f"beer consumption visitors: {correlation_visitors}")
print(f"beer consumption chicken: {correlation_chicken}")
```

```python
import pandas as pd

# Replace 'input.csv' with your CSV file name and 'merged_df.xlsx' with the desired Excel file name
input_csv = 'merged_df.csv'
output_excel = 'merged_df.xlsx'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(input_csv)

# Write the DataFrame to an Excel file
df.to_excel(output_excel, index=False)

```

```python
# VISUALIZATION IN PANDAS 

import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with correlation values
correlation_data = pd.DataFrame({
        'Beer_price': correlation_beer_price,
        'Chicken_consumption': correlation_chicken,
        'Goals': correlation_goals,
          ' Inflation_Rate': correlation_inflation,
          'Visitors_total': correlation_visitors
}, index=['Beer_consumption'])

# Create a heatmap using seaborn
sns.set(style="white")  
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt=".2f")

# Show the plot
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns for which you want to create scatter plots
selected_columns = ['Duration', 'Visitors_total', 'Goals', 'Chicken_consumption', 'Beer_price', ' Inflation_Rate', 'Beer_consumption']

# Subset the DataFrame with the selected columns
selected_df = df[selected_columns]

# Create scatter plots using seaborn's pairplot
sns.set(style="ticks")  # Optional: Set the style
sns.pairplot(selected_df)
plt.suptitle("Pairwise Scatter Plots", y=1.02) 
plt.show()
```

```python
# Select the columns for which you want to create scatter plots
selected_columns = ['Duration', 'Visitors_total', 'Goals', 'Chicken_consumption', 'Beer_price', ' Inflation_Rate', 'Beer_consumption']

# Subset the DataFrame with the selected columns
selected_df = df[selected_columns]

# Set the style (optional)
sns.set(style="whitegrid")

# Loop through each pair of variables and create a scatter plot
for i in range(len(selected_columns)):
    for j in range(i + 1, len(selected_columns)):
        x_col = selected_columns[i]
        y_col = selected_columns[j]
        
        # Create scatter plot
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=x_col, y=y_col, data=selected_df)
        
        # Set plot title
        plt.title(f'Scatter Plot: {y_col} vs {x_col}')
        
        # Show the plot
        plt.show()
```
